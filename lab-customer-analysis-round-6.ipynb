```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
df.head()
```

```python
df.drop(['Customer', 'Effective To Date'], axis = 1, inplace = True)
df.head(3)
```

### - One Hot/Label Encoding (categorical columns).
    Encoding categoricals

    Ordinal: categories have an inherent order (ie: grades) - LabelEncoder
    Nominal: no inherent order (ie: state, gender) - OneHotEncoder


```python
categorical_df = df[df.select_dtypes('object').columns]
categorical_df.head(3)
```

```python
numerical_df = df[df.select_dtypes('number').columns]
numerical_df
```

```python
one_hot_df = pd.get_dummies(categorical_df, drop_first=True)
one_hot_df
```

```python
print(numerical_df.shape)
print(categorical_df.shape)
```

### - Concat DataFrames

```python
df_concat = pd.concat([numerical_df,one_hot_df], axis=1)
df_concat.shape
```

### Linear Regression
- Train-test split.
- Apply linear regression.

```python
#xt: x_total (total.- datos concatenados)
#yt: y_total (total.- datos concatenados)

yt=df_concat['Total Claim Amount']
Xt=df_concat.drop(['Total Claim Amount'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Xt, yt, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as LinReg
linreg=LinReg()                # model
linreg.fit(x_train, y_train)   # model train
y_pred_train = linreg.predict(x_train)
y_pred_test=linreg.predict(x_test) 
```

### Model Validation
Description:
- $R^2$: The coefficient of determination (RÂ²)
- MSE: Mean squared error
- RMSE: Root Mean Squared Error
- MAE: Mean absolute error

*LinearRegression().score(): mÃ©todo de la biblioteca Scikit-learn que se utiliza  
para calcular el coeficiente de determinaciÃ³n (RÂ²) del modelo de regresiÃ³n lineal.*

```python
train_score = linreg.score(x_train, y_train) 
train_score
```

```python
test_score = linreg.score(x_test, y_test)    
test_score
```

La $R^2$ baja un poco para los datos de x_test, ya que estos son los datos guardados

```python
from sklearn.metrics import mean_squared_error as mse
train_mse = mse(y_pred_train,y_train)
test_mse = mse(y_pred_test,y_test)
train_mse, test_mse
```

```python
train_rmse = np.sqrt(train_mse)
test_mse = np.sqrt(test_mse)
train_rmse, test_mse
```

```python
from sklearn.metrics import mean_absolute_error as mae
train_mae = mae(y_pred_train,y_train)
test_mae = mae(y_pred_test,y_test)
train_mae, test_mae
```
